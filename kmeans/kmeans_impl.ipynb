```python
import math
import random as rd
import matplotlib.pyplot as plt

# find the euclidean distance between two points
def eucldist(pt1, pt2):
  return math.sqrt((pt1[0]-pt2[0])**2 + (pt1[1]-pt2[1])**2)

# find the centroid of a list of points
def findCentroid(points):
  xavg = sum(x for x, _ in points)/len(points)
  yavg = sum(y for _, y in points)/len(points)
  return (xavg, yavg)

def plotClusters(clusters, centroids):
  plt.xlim(0,40)
  plt.ylim(0,40)
  colors = ['red','blue','green','yellow']
  i = 0
  for cluster in clusters:
    for pt in cluster:
      plt.plot(pt[0],pt[1],marker="o",markeredgecolor=colors[i],markerfacecolor=colors[i])
    plt.plot(centroids[i][0],centroids[i][1],marker="x",markeredgecolor=colors[i],markerfacecolor=colors[i])
    i = i + 1
  plt.show()

# k - the usual, i.e., the number of clusters
# data - 2d datasets organized into tuples (i.e., pairs)
# maxiterations - if the algorithms hasn't stablized, terminate after maxiterations
# return value: the original data reorganized into k clusters
def kmeans2d(k, data, maxiterations=100):
  # pick k random points as our first cluster centroids
  centroids = rd.sample(data, k)
  print(centroids)
  #alternative native python implementation:
  #centroids = []
  #for _ in range(k):
  #  nextrand = rd.randint(0, len(data))
  #  while nextrand in centroids:
  #    nextrand = rd.randint(0, len(data))
  #  centroids.append(nextrand)

  # (re)associate all points with the closest centroid
  for i in range(maxiterations):
    oldcentroids = centroids.copy()  # need to save old centroids to verify whether centroids haven't changed at all
    clusters = [[] for _ in range(k)] # reset the clusters b/c we will be assigning points to closest cluster again using the new centroids
    for pt in data:
      centroid_dists = [eucldist(c, pt) for c in centroids]
      # - cool way to find index of min val - https://stackoverflow.com/a/13301022 
      _, cidx = min((val, idx) for (idx, val) in enumerate(centroid_dists))
      clusters[cidx].append(pt)
    plotClusters(clusters, centroids)
    for idx,cluster in enumerate(clusters):
      centroids[idx] = findCentroid(cluster)    
    print(i, centroids, clusters)
    # compare centroids with oldcentroids to see if we can quit! 
    if (oldcentroids==centroids):
      break
  return clusters

# let's create random 2d data that is essentially two different clusters
xs = [rd.randint(0,40) for _ in range(50)] + [rd.randint(0,40) for _ in range(50)]
ys = [rd.randint(0,40) for _ in range(100)]
data = list(zip(xs,ys))

clusters = kmeans2d(4, data, 20)
print(clusters)
```
